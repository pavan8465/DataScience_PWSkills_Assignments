1. What are data structures, and why are they important?
- Data structures are ways of organizing and storing data so it can be used efficiently. They're important because they:
  - make programs faster
  - reduce memory usage
  - simplify problem-solving
  
  Examples: list, tuple, dictionary, set.

2. Explain the difference between mutable and immutable data types with examples?
- Mutable: can be changed after creation.

      a = [1, 2, 3]
      a[0] = 10 #can be changed

- Immutable: cannot be changed once created.

      b = (1, 2, 3)
      b[0] = 10 #cannot be changed


3. What are the main differences between lists and tuples in Python?

- Lists are mutable whereas tuples are immutable.
- Lists are slower whereas tuples are faster.
- Data in lists can be changed whereas tuples data cannot be changed.

4. Describe how dictionaries store data?

13. How do sets handle duplicate values in Python?
- Sets store only unique elements. When a set encounters a duplicate value, it simply ignores it.

14. How does the “in” keyword work differently for lists and dictionaries?
- The 'in' keyword checks for the presence of a value in a list, but it checks for the presence of a key in a dictionary.

15. Can you modify the elements of a tuple? Explain why or why not?

17. Describe the time complexity of accessing elements in a dictionary?

18. In what situations are lists preferred over dictionaries?
- List are preferred over dictionaries when

19. Why are dictionaries considered unordered, and how does that affect data retrieval?

- A dictionary stores data as key-value pairs, and internally it uses a hash table. Because of this keys are placed based on their hash value.

20. Explain the difference between a list and a dictionary in terms of data retrieval?
