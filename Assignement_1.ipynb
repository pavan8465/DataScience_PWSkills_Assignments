Q1. What is Python, and why is it popular ?
- Python is a high-level programming language that is easy to read and write. It is popular because it has simple syntax, lots of libraries, and is used in data science, AI, web development, and automation.

Q2.  What is an interpreter in Python ?
- An interpreter runs Python code line by line and shows errors immediately if something goes wrong.

Q3. What are pre-defined keywords in Python?
- Keywords are reserved words in Python like if, else, for, while, True, etc. They have special meanings and cannot be used as a variables.

Q4. Can keywords be used as variable names?
- No, keywords cannot be used as variable names.They are reserved and are for used for specific purposes.

Q5. What is mutability in Python?
- Mutability means an object can be changed/assigned after it is created.

Q6. Why are lists mutable, but tuples are immutable?
- Lists are mutable because they can be changed/ reassigned after they are created but tuples are immutable because they cannot be changed after they are created and are fixed.

Q7. What is the difference between “==” and “is” operators in Python?
- == checks value equality and gives True if values are equal and False if they are not.
- is checks memory location that is if they are same object or not.

Q8. What are logical operators in Python?
- Logical operators are and, or, and not. They are used to combine conditions.

Q9. What is type casting in Python?
- Type casting means converting one data type into another, like string to integer.

Q10.  What is the difference between implicit and explicit type casting ?

- Implicit: It is a type casting in which python converts one data type to another automatically

- Explicit: It is a type casting in which convertion from one data type to another happens manually using int(), float(), etc.

Q11. What is the purpose of conditional statements in Python?

- Conditional statements are those statements which are used to make decisions based on conditions.

Q12. How does the elif statement work?

- Elif is a nested statement of if and will check another condition if the previous 'if' condition is false.

Q13. What is the difference between for and while loops?

- for loop is used when the number of iterations is known and used with range

- while loop is used when you want condition to decide the loop, it can go on forever if conition doesn't become false.

Q14. Describe a scenario where a while loop is more suitable than a for loop?

- When we don't know how many times a loop should run, like waiting for user input until it's correct.

```python
#14. Write a program to reverse a string using a while loop.

s = input("Enter string: ")
i = len(s) - 1

while i >= 0:
    print(s[i], end="")
    i -= 1
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop

num = int(input("Enter number: "))
fact = 1

while num > 0:
    fact *= num
    num -= 1

print("Factorial:", fact)
```
